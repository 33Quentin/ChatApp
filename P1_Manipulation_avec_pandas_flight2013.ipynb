```python
#Maintenant que nous pouvons accéder aux données, nous devons maintenant filtrer pour récupérer uniquement ce qui nous intéresse.
data[data['origin'] == "JFK"].head()
```

```python
#les vols en provenance de l'aéroport JFK le 3 octobre
data[(data['dest'] == "DCA") & (data['month'] == 10) & (data['day'] == 3)]
```

```python
data[(data['hour'] >= 23) | (data['hour'] <= 1)].head()
```

```python
#Récupérer les numéros de série (tailnum) des vols dont l'aéroport de départ est JFK ou EWR.
#Indice : utiliser la fonction .isin() sur la colonne du DataFrame choisie.

data[(data['origin'].isin(['JFK', 'EWR']))]['tailnum']
```

```python
#Supposons que l'on souhaite connaitre le retard enregistré au départ pour tous les vols d'origine JFK et qui ont décollé en novembre 2013.
#Avec l'aide de la fonction .loc, nous pouvons spécifier directement la colonne qui nous intéresse lors du filtrage.
data.loc[(data['origin'] == "JFK") & (data['month'] == 11) & (data['year'] == 2013), 'dep_delay'].head()
```

```python
#Il est ensuite possible de récupérer cette colonne comme un vecteur et de faire des manipulations.
#Dans certains cas, si l'on a besoin de plus d'une colonne, il suffit de spécifier une liste de colonnes à récupérer.
data.loc[(data['origin'] == "JFK") & (data['month'] == 11) & (data['year'] == 2013), ['dep_delay', 'dest']].head()
```

```python
#Afficher l'heure de départ et le retard au départ de tous les vols ayant décollé le 24 décembre 2013.
data.loc[(data['day'] == 24) & (data['month'] == 12) & (data['year'] == 2013), ['dep_time', 'dep_delay']]
```

```python
# D'abord, on récupère uniquement les deux variables
retards = data[['origin', 'dep_delay']]
retards.head()
```

```python
# Ensuite, nous filtrons pour ne récupérer que les retards
retards = retards[retards['dep_delay'] > 0]
retards.head()
```

```python
# Il ne reste plus qu'à grouper et calculer la moyenne
retards.groupby(by='origin').mean()
```

```python
#Déterminer, pour chaque aéroport de destination, le retard le plus important qu'il y ait eu lieu.
retards = data[['dest', 'arr_delay']]
retards.groupby(by='dest').max()
```

```python
#Parmi tous les vols en provenance de l'aéroport JFK, quel est le mois où il y a eu le plus long retard d'un vol à l'arrivée ?
# On ne prends d'abord que ceux dont l'origine est JFK et ensuite, on ne s'intéresse qu'au mois et au retard à l'arrivée
retards = data.loc[data['origin'] == "JFK", ['month', 'arr_delay']]
retards.groupby(by='month').max()
```
