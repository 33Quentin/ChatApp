```python
#Maintenant que nous pouvons acc√©der aux donn√©es, nous devons maintenant filtrer pour r√©cup√©rer uniquement ce qui nous int√©resse.
data[data['origin'] == "JFK"].head()
```

```python
#R√©cup√©rer les num√©ros de s√©rie (tailnum) des vols dont l'a√©roport de d√©part est JFK ou EWR.
#Indice : utiliser la fonction .isin() sur la colonne du DataFrame choisie.

data[(data['origin'].isin(['JFK', 'EWR']))]['tailnum']
```

```python
#Supposons que l'on souhaite connaitre le retard enregistr√© au d√©part pour tous les vols d'origine JFK et qui ont d√©coll√© en novembre 2013.
#Avec l'aide de la fonction .loc, nous pouvons sp√©cifier directement la colonne qui nous int√©resse lors du filtrage.
data.loc[(data['origin'] == "JFK") & (data['month'] == 11) & (data['year'] == 2013), 'dep_delay'].head()
```

```python
#Il est ensuite possible de r√©cup√©rer cette colonne comme un vecteur et de faire des manipulations.
#Dans certains cas, si l'on a besoin de plus d'une colonne, il suffit de sp√©cifier une liste de colonnes √† r√©cup√©rer.
data.loc[(data['origin'] == "JFK") & (data['month'] == 11) & (data['year'] == 2013), ['dep_delay', 'dest']].head()
```

```python
#Afficher l'heure de d√©part et le retard au d√©part de tous les vols ayant d√©coll√© le 24 d√©cembre 2013.
data.loc[(data['day'] == 24) & (data['month'] == 12) & (data['year'] == 2013), ['dep_time', 'dep_delay']]
```

```python
# D'abord, on r√©cup√®re uniquement les deux variables
retards = data[['origin', 'dep_delay']]
retards.head()
```

```python
# Ensuite, nous filtrons pour ne r√©cup√©rer que les retards
retards = retards[retards['dep_delay'] > 0]
retards.head()
```

```python
# Il ne reste plus qu'√† grouper et calculer la moyenne
retards.groupby(by='origin').mean()
```

```python
#D√©terminer, pour chaque a√©roport de destination, le retard le plus important qu'il y ait eu lieu.
retards = data[['dest', 'arr_delay']]
retards.groupby(by='dest').max()
```

```python
#Parmi tous les vols en provenance de l'a√©roport JFK, quel est le mois o√π il y a eu le plus long retard d'un vol √† l'arriv√©e ?
# On ne prends d'abord que ceux dont l'origine est JFK et ensuite, on ne s'int√©resse qu'au mois et au retard √† l'arriv√©e
retards = data.loc[data['origin'] == "JFK", ['month', 'arr_delay']]
retards.groupby(by='month').max()
```
